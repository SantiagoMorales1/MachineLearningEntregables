# Limpieza de los datos <br>

<div style="text-align: justify">
Una vez revisados los datos y sacadas algunas conclusiones la limpieza estará dada a partir de las columnas que contiene valores nulos, por ellos la estrategia será armar un nuevo dataset omitiendo las columnas que tienen muchos registros en nulos o donde la columna se de tipo Object ya que al ser categórica no se pueden utilizar alguna función como máximo, mínimo o promedio para imputar el dato faltante. <br><br>

<div style="text-align: justify">
Finalmente para las columnas donde no hace falta muchos registros y que no son tipo Object estos se imputarán mediante un promedio. De este modo las columnas que se omitirán del nuevo dataset son: MiscFeature, Fence, PoolQC, GarageCond, GarageQual, FireplaceQu, GarageType, GarageFinish, Electrical, BsmtFinType1, BsmtFinType2, BsmtExposure, BsmtCond, BsmtQual, MasVnrType, Alley; y las columnas imputadas serán LotFrontage, MasVnrArea, GarageYrBlt.


```python
#Eliminar las columnas - Tambien se elimina el ID por no ser importante para el analisis
df = df.drop(columns=['MiscFeature', 'Fence', 'PoolQC', 'GarageCond', 'GarageQual', 'FireplaceQu', 'GarageType', 'GarageFinish',
              'Electrical', 'BsmtFinType1', 'BsmtFinType2', 'BsmtExposure', 'BsmtCond', 'BsmtQual', 'MasVnrType', 'Alley', 'Id'])
```

```python
#Reemplazar valor nan

df['LotFrontage'] = df['LotFrontage'].fillna((df['LotFrontage'].mean()))
df['MasVnrArea'] = df['MasVnrArea'].fillna((df['MasVnrArea'].mean()))
df['GarageYrBlt'] = df['GarageYrBlt'].fillna((df['GarageYrBlt'].mean()))
```

```python
#Verificar que no hay valores nulos
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_corr = df.corr()
plt.subplots(figsize=(17,14))
sns.heatmap(df_corr, square=True,vmax=0.9, cmap ='YlGnBu')
```

```python
#Mejores pares de correlacion

corr_pairs = df_corr.unstack()
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
strong_pairs = sorted_pairs[(abs(sorted_pairs) > 0.6) & (abs(sorted_pairs) < 1.0)] 
print(strong_pairs)
```

```python
#Obtener variables categoricas y numericas en datasets diferentes para graficas

VariablesContinuas = []
VariablesCategorica = []

for i in range(len(df.columns)):
    NameCol = (df.columns[i]) 
    
    if (df[NameCol].dtype == 'int64' or df[NameCol].dtype == 'float64'):
        VariablesContinuas.append(NameCol)
    elif (df[NameCol].dtype == 'object'):
        VariablesCategorica.append(NameCol)
    
#print(df[NameCol].dtype)
print (VariablesContinuas)
print (VariablesCategorica)
```

```python
# Graficar en barras las variables categoricas 

import matplotlib.pyplot as plt
import scipy
fig, axarr = plt.subplots(len(VariablesCategorica), 1, figsize=(15,150))

for i in range(0, len(VariablesCategorica)):
    df[VariablesCategorica[i]].value_counts().sort_index().plot.bar(ax=axarr[i])
    axarr[i].set_title(VariablesCategorica[i], fontsize=18)
    fig.tight_layout()
```

```python
import matplotlib.pyplot as plt
import scipy

# Graficar en barras las variables categoricas 
fig, axarr = plt.subplots(len(VariablesContinuas), 1, figsize=(10,150))
red_square = dict(markerfacecolor='r', marker='s')

for i in range(0, len(VariablesContinuas)):
    axarr[i].boxplot(df[VariablesContinuas[i]], vert=False, flierprops=red_square)
    axarr[i].set_title(VariablesContinuas[i], fontsize=18)
    fig.tight_layout()
```
